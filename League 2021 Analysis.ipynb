```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
team_data = df[df['position']=='team']
team_data = team_data.loc[team_data['league'].isin(['LCS','LEC','LCK','LPL','CBLOL','LCL','LJL','LLA','LCO','PCS','TCL','VCS'])]

team_data = team_data[['league','date','playoffs','team','gamelength','result','teamkills','teamdeaths','totalgold']]
team_data = team_data.reset_index()
team_data['enemy_gold'] = None
team_data['enemy_team'] = None
temp_date = None
for index,row in team_data.iterrows():
    if temp_date != row['date']:
        team_data.loc[index,'enemy_gold'] = team_data.loc[index+1,'totalgold']
        team_data.loc[index,'enemy_team'] = team_data.loc[index+1,'team']
        temp_date = row['date']
    else:
        team_data.loc[index,'enemy_gold'] = team_data.loc[index-1,'totalgold']
        team_data.loc[index,'enemy_team'] = team_data.loc[index-1,'team']
        temp_date = row['date']

team_data.rename(columns={'teamkills': 'ally_kills', 'teamdeaths': 'enemy_kills','totalgold':'ally_gold'}, inplace=True)
team_data['gamelength'] = team_data['gamelength']/60
team_data.insert(0, 'elo_rating', 1000)
team_data.insert(3, 'new_elo',None)
team_data.insert(0, 'gamenum', 1)
team_data['enemy_elo'] = 1000
team_data['ally_gold'] = round(team_data['ally_gold'] / team_data['gamelength'],2) 
team_data['enemy_gold'] = team_data['enemy_gold'] / team_data['gamelength']
team_data.insert(9,'gold_diff', team_data['ally_gold']-team_data['enemy_gold'])
team_data.insert(10,'kill_diff', team_data['ally_kills']-team_data['enemy_kills'])

team_data.head(10)
```

```python
kill_diff_std = abs(team_data['kill_diff']).std()
gold_diff_std = abs(team_data['gold_diff']).std()

team_dict = {}
elo_dict = {}
temp_date = None
for index, row in team_data.iterrows():
    if row['team'] in team_dict:
        team_dict[row['team']] = team_dict[row['team']] + 1
        team_data.loc[index,'gamenum'] = team_dict[row['team']] 
    else:
        team_dict[row['team']] = row['gamenum']

for index, row in team_data.iterrows():        
    ##only runs if previous row wasn't the same match
    if temp_date != row['date']:
        ##checks if team occurred previously (stored in dictionary)
        if row['team'] in elo_dict:
            ##checks if enemy team occurred previously and takes the elo (stored in dictionary)
            if row['enemy_team'] in elo_dict:
                enemy_elo = elo_dict[row['enemy_team']]
            ##else grabs it from row (which will be the initial elo of 1000)
            else:
                enemy_elo = row['enemy_elo']
            ##since team is in dictionary (implying it is updated from base) takes updated elo of that team    
            elo_rating = elo_dict[row['team']]
        ##if team isn't in dict but enemy is, takes team elo from row data (1000) and enemy from dict
        elif row['enemy_team'] in elo_dict:
            elo_rating = row['elo_rating']
            enemy_elo = elo_dict[row['enemy_team']]
        ##if neither in dict, takes from row (1000)
        else:
            elo_rating = row['elo_rating']
            enemy_elo = row['enemy_elo']
            
        ##getting dictionary data and replacing old values
        team_data.loc[index,'elo_rating'] = elo_rating
        team_data.loc[index,'enemy_elo'] = enemy_elo

        ##calculating elo and putting it in dictionary and new_elo value
        dr = elo_rating - enemy_elo
        W_e = 1 / (10**(-(float(dr))/400) + 1)
        if row['playoffs'] == 1:
            elo_dict[row['team']] = row['elo_rating'] + 150*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
        else:
            elo_dict[row['team']] = row['elo_rating'] + 100*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
#         elo_dict[row['team']] = row['elo_rating'] + 100*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
#         elo_dict[row['team']] = row['elo_rating'] + 40*(1+abs(row['gold_diff'])/gold_diff_std)*(row['result']-W_e)        
        team_data.loc[index,'new_elo'] = elo_dict[row['team']]
        
        ##grabbing date of match to ensure duplication is taken account of
        temp_date = row['date']
    ##if date is same, swaps the elo from previous loop
    else:
        ##replacing old values of swap (current team was enemy in previous loop, and current enemy was team)
        team_data.loc[index,'elo_rating'] = enemy_elo
        team_data.loc[index,'enemy_elo'] = elo_rating
        ##swapping holder variable values
        elo_rating =  team_data.loc[index,'elo_rating']
        enemy_elo = team_data.loc[index,'enemy_elo']
        
        ##calculating elo and putting it in dictionary and new_elo value
        dr = elo_rating - enemy_elo
        W_e = 1 / (10**(-(float(dr))/400) + 1) 
        if row['playoffs'] == 1:
            elo_dict[row['team']] = row['elo_rating'] + 150*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
        else:
            elo_dict[row['team']] = row['elo_rating'] + 100*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
#         elo_dict[row['team']] = row['elo_rating'] + 100*(1+abs(row['gold_diff'])/gold_diff_std+abs(row['kill_diff'])/kill_diff_std)*(row['result']-W_e)
#         elo_dict[row['team']] = row['elo_rating'] + 40*(1+abs(row['gold_diff'])/gold_diff_std)*(row['result']-W_e)        
        team_data.loc[index,'new_elo'] = elo_dict[row['team']]
        
        ##grabbing date of match to ensure duplication is taken account of
        temp_date = row['date']
        
        
team_data.head()
```
